```python
def dist(X, Y):
    sx = np.sum(X**2, axis=1, keepdims=True)
    sy = np.sum(Y**2, axis=1, keepdims=True)
    return np.sqrt(-2 * X.dot(Y.T) + sx + sy.T)
```

```python
def precision(predictions):
    true_positive = np.sum((predictions == test_np_dep) & (test_np_dep == 1))
    false_positive = np.sum((predictions != test_np_dep) & (test_np_dep == 0))
    precision = true_positive / (true_positive + false_positive)
    return precision
```

```python
def recall(predictions):
    true_positive = np.sum((predictions == test_np_dep) & (test_np_dep == 1))
    false_negative = np.sum((predictions != test_np_dep) & (test_np_dep == 1))
    recall = true_positive / (true_positive + false_negative)
    return recall
```

```python
def normalize(x):
    x_norm = (x - np.min(x)) / (np.max(x) - np.min(x))
    return x_norm
```
